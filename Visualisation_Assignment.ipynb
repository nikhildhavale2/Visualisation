#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Generate synthetic dataset
np.random.seed(0)
data = np.random.multivariate_normal([0, 0], [[1, 0.8], [0.8, 1]], size=100)
dataset = pd.DataFrame(data, columns=['Feature_1', 'Feature_2'])

# Compute the correlation matrix
corr = dataset.corr()

# Generate a heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix Heatmap")
plt.show()
```

#3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.

np.random.seed(20)

data = {

    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),

    'Day': np.random.choice(range(1, 31), 100),

    'Sales': np.random.randint(1000, 5000, 100)

df = pd.DataFrame(data)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sales data
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot the data to get a matrix format using pivot_table to handle duplicates
# aggfunc='sum' will sum the sales for duplicate Day-Month combinations
sales_pivot = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='sum')

# Create heatmap
# Changed fmt to '.0f' to display floats as integers without decimal places
sns.heatmap(sales_pivot, annot=True, fmt='.0f', cmap='viridis')
plt.title("Sales Variation Across Different Months and Days")
plt.xlabel("Month")
plt.ylabel("Day")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd

# Generate the x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calculate z based on the given function
z = np.sin(np.sqrt(x**2 + y**2))

# Prepare the data for a pandas DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z, cmap='viridis')

# Add labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Show the plot
plt.show()
```

#5. Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Given data
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Plot Population and GDP
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
df.plot(kind='bar', x='Country', y='Population', ax=ax1, color='b', position=0, width=0.4)
df.plot(kind='bar', x='Country', y='GDP', ax=ax2, color='r', position=1, width=0.4)

ax1.set_xlabel('Country')
ax1.set_ylabel('Population', color='b')
ax2.set_ylabel('GDP', color='r')

plt.title("Comparison of Population and GDP by Country")
plt.show()
```

#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, show, output_notebook
import numpy as np

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Output to notebook
output_notebook()

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="Sine", line_width=2)

# Show the results
show(p)
```

#2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import push_notebook
from bokeh.models import ColumnDataSource
import numpy as np

# Generate random data
np.random.seed(42)
x = np.random.random(100) * 100
y = np.random.random(100) * 100
sizes = np.random.random(100) * 30
colors = np.random.random(100) * 255

# Prepare the data
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Output to notebook
output_notebook()

# Create a new plot
p = figure(title="Scatter Plot with Variable Sizes and Colors", x_axis_label='x', y_axis_label='y')

# Add scatter renderer
p.scatter('x', 'y', size='sizes', color='colors', fill_alpha=0.6, source=source, legend_label="Random Data")

# Show the results
show(p, notebook_handle=True)
```

#3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']

counts = [20, 25, 30, 35]



```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import push_notebook

# Data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Output to notebook
output_notebook()

# Create a new plot
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count')

# Add bar renderer
p.vbar(x=fruits, top=counts, width=0.5)

# Show the results
show(p, notebook_handle=True)
```

#4. Create a Bokeh histogram to visualize the distribution of the given data.

data_hist = np.random.randn(1000)

hist, edges = np.histogram(data_hist, bins=30)

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import push_notebook
import numpy as np

# Given data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Output to notebook
output_notebook()

# Create a new plot
p = figure(title="Histogram of Data", x_axis_label='Value', y_axis_label='Frequency')

# Add quad glyphs to create histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.7)

# Show the results
show(p, notebook_handle=True)
```

#5. Create a Bokeh heatmap using the provided dataset.

data_heatmap = np.random.rand(10, 10)

x = np.linspace(0, 1, 10)

y = np.linspace(0, 1, 10)

xx, yy = np.meshgrid(x, y)

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import push_notebook
from bokeh.models import LinearColorMapper, ColorBar
import numpy as np

# Provided data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Output to notebook
output_notebook()

# Create a new plot
p = figure(title="Heatmap", x_axis_label='x', y_axis_label='y')

# Define color mapper
color_mapper = LinearColorMapper(palette='Viridis256', low=data_heatmap.min(), high=data_heatmap.max())

# Add image renderer
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

# Add color bar
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=12, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the results
show(p, notebook_handle=True)
```
